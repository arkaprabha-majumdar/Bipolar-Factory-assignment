```
import pandas as pd
import numpy as np
import nltk
nltk.download('words')
from nltk.corpus import words
```

```
'World'.lower() in words.words()
```

```
import requests
from bs4 import BeautifulSoup
def fetch(url):
    r1 = requests.get(url)
    coverpage = r1.content
    soup1 = BeautifulSoup(coverpage, 'html5lib')
    list_links = soup1.find_all('a', class_='')
    return list_links
```

```
def article_links(url):
    if len(url.get_text())>25: #removes waste links like "contact us","home", etc
      return True
    else:
      return False
```

```
url=input(prompt="Enter your news site  : ") or "https://www.foxnews.com" #default url i used
```

```
def separate(url): #separates article links and other links like "/music" or "/entertainment"
  articles=[]       
  others=[]
  c=0
  for link in fetch(url):
    print(c)
    if len(link.get_text())>35and len(link['href'])>45:
      articles.append(link['href'])
    else:
      if link.get_text().lower() in words.words():
        others.append(link['href'])
    c+=1
  return articles,others
```

```
articles,others=separate(url)
articles=[article for article in articles if len(article)>5]
others=[article for article in others if len(article)>5]
```

```
def http_conv(linklist):
  for i in range(len(linklist)):
    if 'http' not in linklist[i]:
      linklist[i]='https:'+linklist[i]
  return linklist #adds http:// to beginning of link if not there
```

```
articles=http_conv(articles)
others=http_conv(others)
```

```
crawled_list=[]
counter=0
for url in others:
  print(counter)
  crawled_list.append(fetch(url))
  counter+=1 '''for each of the other links, say "/music", "/sports", i re fetched the urls from each of the pages, boosting our potential links from 
                150 to ~21000'''
```

```
import itertools
crawled_list  = list(itertools.chain(*crawled_list)) #flattens the list of list to one list
```

```
crawl=[]
for item in crawled_list:
  try:
    crawl.append(item['href'])
  except:
    TypeError
    pass
crawled_list=http_conv(crawl)
```
